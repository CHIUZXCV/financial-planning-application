# Monte Carlo Simulation

In this step, you will run Monte Carlo Simulations for your portfolio to model portfolio performance at different retirement ages. 

Complete the following steps:
1. Calculate the daily returns for the SPY and AGG closing prices.
2. Calculate volatility for both the SPY and AGG closing prices.
3. Find the last day's closing price for both stocks and save those as variables.
4. Run a Monte Carlo Simulation of at least 500 iterations and generate at least 30 years of closing prices

### HINTS:
There are 252 trading days per year, so the number of records to generate for each Monte Carlo run will be 252 days * 30 years

```python
# Calculate the daily roi for the stocks
daily_returns = df.pct_change()
daily_returns
```

```python
avg_daily_returns_spy = daily_returns.mean()["SPY"]['close']
avg_daily_returns_agg = daily_returns.mean()["AGG"]['close']
```

```python
# Calculate volatility
std_daily_returns_spy = daily_returns.std()["SPY"]['close']
std_daily_returns_agg = daily_returns.std()["AGG"]['close']
```

```python
# Save the last day's closing price
last_closing_spy = df['SPY']['close'][-1]
last_closing_agg = df['AGG']['close'][-1]
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 500
number_records = 252 * 30
monte_carlo = pd.DataFrame()
portfolio_cumulative_returns = pd.DataFrame()

#spy_randoms = np.random.normal(avg_daily_returns_spy,std_daily_returns_spy)
#agg_randoms = np.random.normal(avg_daily_returns_agg,std_daily_returns_agg)
```

```python
# Run the Monte Carlo Simulation
for x in range(number_simulations):
    simulated_spy_prices = [last_closing_spy]
    simulated_agg_prices = [last_closing_agg]
   
    

    for i in range(number_records):
        simulated_Spy_price = simulated_spy_prices[-1] * (1 + np.random.normal(avg_daily_returns_spy,std_daily_returns_spy))
        simulated_agg_price = simulated_agg_prices[-1] * (1 + np.random.normal(avg_daily_returns_agg,std_daily_returns_agg))
        
        simulated_spy_prices.append(simulated_Spy_price)
        simulated_agg_prices.append(simulated_agg_price)
       

        
    monte_carlo["SPY prices"] = pd.Series(simulated_spy_prices)
    monte_carlo["AGG prices"] = pd.Series(simulated_agg_prices)

    simulated_daily_returns = monte_carlo.pct_change()

    weights = [0.6, 0.4]

    portfolio_daily_returns = simulated_daily_returns.dot(weights)

    portfolio_cumulative_returns[x] = (1 + portfolio_daily_returns.fillna(0)).cumprod()

portfolio_cumulative_returns.head()
```

```python
# Visualize the Simulation
portfolio_cumulative_returns.plot(legend = None, figsize= (20,10))
```

```python
# Select the last row for the cumulative returns (cumulative returns at 30 years)
cumulative_returns_at_30_years= portfolio_cumulative_returns.iloc[-1,:]
cumulative_returns_at_30_years
```

```python
# Select the last row for the cumulative returns (cumulative returns at 20 years)
# YOUR CODE HERE
cumulative_returns_at_20_years = portfolio_cumulative_returns.iloc[5040,:]
cumulative_returns_at_20_years
```

```python
# Display the 90% confidence interval for the ending returns
# YOUR CODE HERE
confidence_interval_30_years = cumulative_returns_at_30_years.quantile(q=[0.05,0.95])
confidence_interval_20_years = cumulative_returns_at_20_years.quantile(q=[0.05,0.95])
confidence_interval_30_years
```

```python
# Visualize the distribution of the ending returns
# YOUR CODE HERE
plt.figure();
cumulative_returns_at_30_years.plot(kind = 'hist', bins = 20, title = "30 years", density = True, figsize = (20,10))
plt.axvline(confidence_interval_30_years.iloc[0], color = 'r')
plt.axvline(confidence_interval_30_years.iloc[1], color = 'r')
```

```python
plt.figure();
cumulative_returns_at_20_years.plot(kind = 'hist', bins = 20, title = "20 years", density = True, figsize = (20,10))
plt.axvline(confidence_interval_20_years.iloc[0], color = 'r')
plt.axvline(confidence_interval_20_years.iloc[1], color = 'r')
```
